```python
import torch
import torch.nn as nn
# 모델 최적화하는 데 필요한 최적화 함수들이 들어가 있음(SGD)
import torch.optim as optim
# 텐서를 초기화하는 함수들이 들어가 있음(normal distribution)
import torch.nn.init as init
import torchvision
import matplotlib.pyplot as plt
```

## data making

```python
# 갯수는 임의 지정
num_data = 1000
num_epoch = 500

# 노이즈 지정(평균은 디폴트 0)
noise = init.normal_(torch.FloatTensor(num_data, 1), std = 0.2); noise.shape
```

```python
noise.numpy().shape
```

```python
# x는 -10에서 10 uniform
x = init.uniform_(torch.Tensor(num_data, 1), -10, 10); x.shape
```

```python
x.numpy().shape
```

```python
# 연산 그래프 정의
y = 2 * x + 3
```

```python
# y에 노이즈를 더해주자
# 실제 데이터에서 노이즈가 추가되는 경우가 많음
y_noise = y + noise
```

## data visualization

```python
plt.figure(figsize = (10, 10))
# 나타내기 위해 numpy로 변환, 점 크기는 7
plt.scatter(x.numpy(), y_noise.numpy(), s = 7, c =  'gray')
# [xmin xmax ymin ymax]
plt.axis([-12, 12, -25, 25])
plt.show()
```

## model, loss function, minimize function

```python
# 입력이 x = (1000, 1), 
# y = (1000, 1)이라서 특성의 갯수는 하나라 Linear(x[1], y[1])
model = nn.Linear(1, 1)

# 손실 함수 지정
# 임의로 L1 사용
loss_func = nn.L1Loss()

# 최적화 함수 지정
# 인수로 학습의 대상이 되는 모델의 변수 전달
# 학습률은 임의
optimizer = optim.SGD(model.parameters(), lr = 0.01)
```

## model training

```python
# 손실이 어떻게 변하는지 확인하기 위해 loss_arr을 만들어 기록
loss_arr = []

# target은 y_noise
label = y_noise

# 500번 학습

for i in range(num_epoch):
    
    # 이전 기울기 지우고 최적화 함수 초기화
    # 기울기를 지우지 않으면 학습이 잘 되지 않음(그때그때 구해야 함)
    optimizer.zero_grad()
    
    # 입력값 x를 모델에 넣어 결과값을 얻음
    output = model(x) # nn.Linear(1, 1)(x)
    
    # 결과값과 목표값의 차이를 L1 손실 함수로 구해줌
    loss = loss_func(output, label)
    
    # 손실에 대한 기울기를 구함
    loss.backward()
    
    # 구한 기울기를 이용해 모델의 변수를 업데이트
    optimizer.step()
    
    
    # 10번마다 모델의 변수가 어떻게 변하고 있는지 출력해줌
    if i % 10 == 0:
        
        plt.scatter(x.detach().numpy(), output.detach().numpy())
        plt.axis([-10, 10, -30, 30])
        plt.show()
        params = list(model.parameters())
        W = params[0].item()
        b = params[1].item()
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
            i, num_epoch, W, b, loss.data))
    loss_arr.append(loss.detach().numpy())
```

```python
plt.figure(figsize=(15,15))
plt.scatter(x.numpy(),y_noise.numpy(),s=5,c="gray")
plt.scatter(x.detach().numpy(),output.detach().numpy(),s=5,c="red")
plt.axis([-10, 10, -30, 30])
plt.show()
```

## loss graph

```python
plt.plot(loss_arr)
plt.show()
```
